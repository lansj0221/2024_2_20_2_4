```python
import numpy as np
import pandas as pd
from pandas import Series

scores = np.random.randint(50,101,size=(50,5))
scores_df = pd.DataFrame(scores,columns=['國文','英文','數學','地理','歷史'],index=range(1,51))

def get_level(s:Series):
    average = s.mean()
    if average < 60:
        return '不及格'
    elif average < 80:
        return '甲'
    else:
        return '優'

def countfail(s:Series):
    failsubject = s[s < 60]
    countsubject = failsubject.count()
    return countsubject

def main():
    scores_sum = scores_df.sum(axis=1)
    scores_average = scores_df.mean(axis=1)
    scores_df['總分'] = scores_sum
    scores_df['平均'] = scores_average
    scores_df['等級'] = scores_df[['國文','英文','數學','地理','歷史']].apply(get_level,axis=1)
    scores_df['不及格科目數'] = scores_df[['國文','英文','數學','地理','歷史']].apply(countfail,axis=1)
    display(scores_df)

main()
```
